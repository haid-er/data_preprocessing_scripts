# Plot After PCA

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# List of file paths
file_paths = [
    r'F:\ServerData\jan21\Smart_Glass\test\random\test_ACCELEROMETER_e0.csv',
    r'F:\ServerData\jan21\Smart_Phone\test\random\test_Linear Acceleration Sensor_e0.csv',
    r'F:\ServerData\jan21\Smart_Watch\test\random\test_LSM6DSO Accelerometer_e0.csv',
]

# List to store DataFrames
dfs = []

# List to store reduced data for each file
reduced_dfs = []

for file_path in file_paths:
    # Read the file into a DataFrame with space as the delimiter
    df = pd.read_csv(file_path, delimiter=",")
    
    # Add column names for files with 4 columns
    df.columns = ['timestamp', 'x', 'y', 'z']
    
    # Perform dimension reduction using PCA
    n_components = 2
    pca = PCA(n_components=n_components)
    reduced_data = pca.fit_transform(df[['x', 'y', 'z']])
    
    # Create a new DataFrame with the reduced data
    reduced_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])
    
    # Append the reduced DataFrame to the list
    reduced_dfs.append(reduced_df)

# Plot the reduced data for all files in a single graph
plt.figure(figsize=(10, 6))
labels = ['BITF21M501','BITF21M516','BITF21M526','BSEF21M521', 'Imran','xyz']
for i, reduced_df in enumerate(reduced_dfs):
    #plt.plot(reduced_df['PC1'], label= labels[i])
    plt.plot(reduced_df['PC2'], label=labels[i])
    #break

plt.title('Sitting Activity for Smart Glasses Accelerometer')
plt.xlabel('Time')
plt.ylabel('Principal Components')
plt.legend()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# List of file paths
file_paths = [
    r'C:\Users\Dr Hassan.DESKTOP-HJCCLKQ\Downloads\Server_Data_Folder\Smart_Glass\BITF21M501\sitting\BITF21M501_ACCELEROMETER.csv',
    r'C:\Users\Dr Hassan.DESKTOP-HJCCLKQ\Downloads\Server_Data_Folder\Smart_Glass\BITF21M516\sitting\BITF21M516_ACCELEROMETER.csv',
    r'C:\Users\Dr Hassan.DESKTOP-HJCCLKQ\Downloads\Server_Data_Folder\Smart_Glass\BITF21M526\sitting\BITF21M526_ACCELEROMETER.csv',
    r'C:\Users\Dr Hassan.DESKTOP-HJCCLKQ\Downloads\Server_Data_Folder\Smart_Glass\BSEF21M521\sitting\BSEF21M521_ACCELEROMETER.csv',
    r'C:\Users\Dr Hassan.DESKTOP-HJCCLKQ\Downloads\Server_Data_Folder\Smart_Glass\Imran\sitting\Imran_ACCELEROMETER.csv',
    r'C:\Users\Dr Hassan.DESKTOP-HJCCLKQ\Downloads\Server_Data_Folder\Smart_Glass\xyz\sitting\xyz_ACCELEROMETER.csv'
]

# List to store magnitudes for each file
magnitudes = []

for file_path in file_paths:
    # Read the file into a DataFrame with comma as the delimiter
    df = pd.read_csv(file_path, delimiter=",")

    # Add column names for files with 4 columns
    df.columns = ['timestamp', 'x', 'y', 'z']

    # Calculate the magnitude of the accelerometer data
    df['magnitude'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)

    # Append the magnitude to the list
    magnitudes.append(df['magnitude'])

# Plot the magnitude data for all files in a single graph
plt.figure(figsize=(10, 6))
labels = ['BITF21M501', 'BITF21M516', 'BITF21M526', 'BSEF21M521', 'Imran', 'xyz']
for i, magnitude in enumerate(magnitudes):
    plt.plot(magnitude.values, label=labels[i])

plt.title('Sitting Activity for Smart Glasses Accelerometer (Magnitude)')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.legend()
plt.show()
```
